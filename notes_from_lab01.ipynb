# How to make a plot with myplotlib.pyplot

Let's create 1000 random values:

```python
import numpy as np
import matplotlib.pyplot as plt

rand_x = np.random.normal(size=1000)
```

Let's look what rand_x is:

```python
rand_x.shape
```

Let's look inside:

```python
print(rand_x)
```

It is a list of 1000 elements.

```python
# Returns a range of integers from 0 to passed integer parameter:
range(rand_x.shape[0])
```

Now let's try to plot them:

```python
# Let's define the plot size (done separately for each plot)
plt.figure(figsize=(15, 5))
# Here we pass two parameters:
# first - x - range(rand_x.shape), this is list of numbers from 0 to 1000
# second - y - generated random 1000 numbers
plt.scatter(range(rand_x.shape[0]), rand_x)
```

Another plot option:

```python
plt.figure(figsize=(15, 5))
plt.plot(rand_x)
```

# How to make a plot with altair

Let's load the dataset:

```python
# load an example dataset
from vega_datasets import data
cars = data.cars()
```

Let's check how the data looks like:

```python
cars.head(3)
```

```python
# plot the dataset, referencing dataframe column names
import altair as alt
# Enocde maps columns to visual attributes
# Mark_point() says that the atributes should be representes as points
# Interactive allows to interact with the plot
alt.Chart(cars).mark_point().encode(
  x='Horsepower',
  y='Miles_per_Gallon',
  color='Origin'
).interactive()
```

# How to make histogram

```python
# Rand_x contains only list of numbers, so in order to print the histogram
# we need to do the binning of the values (divide the values to the categories)
plt.hist(rand_x, bins=50)
plt.xlabel('values')
plt.ylabel('no of values')
plt.show()

print('min:{:.4}, max:{:.4}, mean:{:.4}, std:{:.4}'.format(rand_x.min(), rand_x.max(), rand_x.mean(), rand_x.std()))
```

We can use more bins:

```python
plt.hist(rand_x, bins=5000)
plt.xlabel('values')
plt.ylabel('no of values')
plt.show()
```

#Numpy Basics

Creating array:

```python
import numpy as np

# create ndarray
inp = [[1, 2],[3, 4]]
X = np.array(inp)
```

Printing values/properties:

```python
print(X)
```

```python
print(X.ndim)
```

```python
print(X.shape)
```

```python
print(X.shape[0])
```

```python
print(type(X))
```

```python
print(X.dtype.name)
```

```python
print(X[0,0])
```

Print first row:

```python
print(X[0,:])
```

Print first column:

```python
print(X[:,0])
```

Printing with formatting:

```python
print(f'1st column: { X[:, 0]     }\n')
```

Create array of ones:

```python
X = np.ones((2, 3), dtype=np.float32)
print('\nX:\n', X)
```

Create array of random integers:

```python
Y = np.random.randint(low=0, high=10, size=(3, 2))
print('\nY:\n', Y)
```

Create another array of random integers:

```python
Z = np.random.randint(low=0, high=10, size=(2, 3))
print('\nZ:\n', Z)
```

Transosition of array:

```python
# transposition, equivalent Z.transpose() or np.transpose(Z)
print('\nZ\':\n', Z.T)  
```

```python
print('\nZ\':\n', Z.transpose())  
```

```python
print('\nZ\':\n', np.transpose(Z))  
```

Matrix addition:

```python
print('\nX:\n', X)
print('\nZ:\n', Z)
A = X + Z
print('\nA:\n', A)
```

Addition of constant to the matrix:

```python
print('\nZ:\n', Z)
B = 3 + Z
print('\nB:\n', B)
```

Multiplying matrix with number:

```python
print('\nZ:\n', Z)
# broadcasting:
C = 2 * Z
print('\nC:\n', C)
```

```python
# tmp = C - 2*Z
# D = tmp.sum()

E = (C - 2. * Z)
print('\nE:\n', E)

D = E.sum()
print('\nD:\n', D)

print('\nZ.sum():\n', Z.sum())
```
