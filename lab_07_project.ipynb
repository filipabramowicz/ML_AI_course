```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/GokuMohandas/MadeWithML/main/datasets/titanic.csv')
df.head(n=20)
```

```python
df.drop(['name'], axis='columns', inplace=True)
```

```python
import numpy as np
df['sex'] = np.where(df['sex'] == 'male', 1, 0)
```

```python
df[df['sex'].isnull()]
```

```python
df['age'].isna().sum()
```

```python
from sklearn.impute import KNNImputer
knn = KNNImputer()

columns = ['age']
age_for_imputing = pd.DataFrame(data=df['age'], columns=columns)
df['age'] = knn.fit_transform(age_for_imputing)
df['age'].isna().sum()
```

```python
df.drop(['ticket'], axis='columns', inplace=True)
```

```python
df[df['fare'].isnull()]
```

```python
columns = ['fare']
age_for_imputing = pd.DataFrame(data=df['fare'], columns=columns)
df['fare'] = knn.fit_transform(age_for_imputing)
df['fare'].isna().sum()
```

```python
df.drop(['cabin'], axis='columns', inplace=True)
```

```python
print(df['embarked'].isna().sum())
```

```python
df.dropna(subset=['embarked'], inplace=True)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns = ['embarked'])
```

```python
df.head()
```

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()

df2 = pd.DataFrame(scaler.fit_transform(df[['pclass','sex','age','sibsp','parch','fare','embarked_C','embarked_Q','embarked_S']]),columns=['pclass','sex','age','sibsp','parch','fare','embarked_C','embarked_Q','embarked_S'])
df2.head()
```

```python
from sklearn.model_selection import train_test_split
input_data = df2[['pclass','sex','age','sibsp','parch','fare','embarked_C','embarked_Q','embarked_S']]
output_data = df['survived']
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.3, random_state=33)
```

```python
import tensorflow as tf
from tensorflow.keras.layers import Dense
from tensorflow.keras import Sequential

model = tf.keras.Sequential()
model.add(Dense(5, activation='relu', input_dim=9, name="L1"))
model.add(Dense(3, activation='relu', name="L2"))
model.add(Dense(1, activation='sigmoid', name="L3"))

model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.001),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

model.summary()
```

```python
model.fit(x_train, y_train, 
          epochs=100, 
          batch_size=16, 
          verbose=1)
```

```python
y_pred = model.predict(x_test)
```

```python
  from sklearn import metrics
  print(f'MAE: {metrics.mean_absolute_error(y_test, y_pred):.2f}')
  print(f'MSE: {metrics.mean_squared_error(y_test, y_pred):.2f}')
```
